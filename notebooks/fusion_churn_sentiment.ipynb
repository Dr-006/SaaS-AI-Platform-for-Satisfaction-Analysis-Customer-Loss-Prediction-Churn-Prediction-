```python
# Cellule 1

import pandas as pd
import joblib
```

```python
# Cellule 2 — Charger dataset churn

df_churn = pd.read_csv("C:/Users/REDA/ai-saas-platform/data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("Shape:", df_churn.shape)
df_churn.head()
```

```python
# Cellule 3 — Générer texte (Feature Engineering)

def generate_fake_feedback(row):
    if row["MonthlyCharges"] > 80:
        return "Too expensive, not satisfied"
    elif row["tenure"] > 24:
        return "Very happy with the service"
    elif row["Contract"] == "Month-to-month":
        return "Service is flexible but could be better"
    else:
        return "Service is okay"

df_churn["text"] = df_churn.apply(generate_fake_feedback, axis=1)

df_churn[["MonthlyCharges", "tenure", "Contract", "text"]].head()
```

```python
# Cellule 4 — Charger modèle NLP

sentiment_model = joblib.load("C:/Users/REDA/ai-saas-platform/backend/app/ml/models/sentiment_model.pkl")
vectorizer = joblib.load("C:/Users/REDA/ai-saas-platform/backend/app/ml/models/vectorizer.pkl")

print("Modèles chargés avec succès")
```

```python
# Cellule 5 — Transformer texte → vecteurs

X_text = vectorizer.transform(df_churn["text"])

print("Transformation terminée")
```

```python
# Cellule 6 — Générer sentiment 

# df_churn["sentiment_score"] = sentiment_model.predict(X_text)
```

```python
#  Remplace la cellule 6 par celle-ci si tu veux un modèle plus puissant :

df_churn["sentiment_score"] = sentiment_model.predict_proba(X_text)[:, 1]
```

```python
# Cellule 8 — Vérification

print(df_churn[["text", "sentiment_score"]].head(10))

print("\nDistribution sentiment :")
print(df_churn["sentiment_score"].value_counts())
```

```python
# Cellule 9 — Sauvegarde dataset fusionné

df_churn.to_csv("C:/Users/REDA/ai-saas-platform/data/processed/churn_with_sentiment.csv", index=False)

print("Fichier sauvegardé avec succès ✅")
```
